```python
import json
import pandas as pd
import plotly.express as px
from urllib.request import urlopen
```

```python
df = pd.read_csv('Employee_Salaries.csv')
salario = pd.DataFrame(df)

salario
```

```python
salario = salario.rename(columns={'Department': 'Departamento', 'Department_Name': 'Nome_Do_Departamento', 'Division': 'Divisão', 'Gender': 'Gênero', 'Base_Salary': 'Base_Salarial', 'Overtime_Pay': 'Pagamento_de_Hora_Extra',
                                  'Longevity_Pay': 'Pagamento_por_Longevidade', 'Grade': 'Nota'})

salario
```

```python
salario['Salario_Total'] = salario['Base_Salarial'] + salario['Pagamento_de_Hora_Extra'] + salario['Pagamento_por_Longevidade']

salario
```

```python
salario['Salario_Total'].sum()
```

```python
def format_number(value, prefix = ''):
    for unit in ['', 'mil', 'milhões', 'bilhões']:
        if value < 1000:
            return f'{prefix} {value:.3f} {unit}'
        value /= 1000
    return f'{prefix} {value:.3f} trilhões'
```

```python
total_salarios = salario['Salario_Total'].sum()
total_salarios
```

```python
salario['Salario_Total'] = salario['Salario_Total'].apply(lambda x: f'${x}')

print(salario['Salario_Total'])
```

```python
salario['Salario_Total'] = salario['Salario_Total'].str.replace('$', '').str.replace(',', '').astype(float)

soma_salarios = salario.groupby('Nome_Do_Departamento')['Salario_Total'].sum()

soma_salarios_formatado = soma_salarios.apply(lambda x: '${:,.2f}'.format(x))

print(soma_salarios_formatado)
```

```python
df_salario_departamento = salario.groupby('Nome_Do_Departamento')[['Salario_Total']].sum()
df_salario_departamento = salario.drop_duplicates(subset='Nome_Do_Departamento')[['Nome_Do_Departamento']].merge(df_salario_departamento, left_on='Nome_Do_Departamento', right_index=True).sort_values('Salario_Total', ascending=False)
print(df_salario_departamento)
```

```python
top7_departamentos = df_salario_departamento.nlargest(7, 'Salario_Total')

top7_departamentos = top7_departamentos.sort_values('Salario_Total')

grafico_salarios_por_departamento = px.bar(
    top7_departamentos,
    x='Nome_Do_Departamento',
    y='Salario_Total',
    title='Salário Total dos 7 Melhores Departamentos',
)

grafico_salarios_por_departamento.show()
```

```python
soma_salarios_por_genero = salario.groupby('Gênero')['Salario_Total'].sum().reset_index()

grafico_salarios_por_genero = px.pie(
    soma_salarios_por_genero,
    names='Gênero',
    values='Salario_Total',
    title='Total de Salários por Gênero',
)

grafico_salarios_por_genero.show()
```

```python
df_horaextra_departamento_genero = salario.groupby(['Nome_Do_Departamento', 'Gênero'])[['Pagamento_de_Hora_Extra']].sum().reset_index()

print(df_horaextra_departamento_genero)
```

```python
top5_departamentos = df_horaextra_departamento_genero.nlargest(10, 'Pagamento_de_Hora_Extra').head(10)

grafico_df_horaextra_departamento = px.bar(
    top5_departamentos,
    x='Nome_Do_Departamento',
    y='Pagamento_de_Hora_Extra',
    color='Gênero',
    title='Pagamento por Hora Extra dos 6 Melhores Departamentos',
    barmode='relative',
)


grafico_df_horaextra_departamento.show()
```

```python
df_longevidade_departamento_genero = salario.groupby(['Nome_Do_Departamento', 'Gênero'])[['Pagamento_por_Longevidade']].sum().reset_index()

print(df_longevidade_departamento_genero)
```

```python
top_departamentos_longevidade = df_longevidade_departamento_genero.nlargest(6, 'Pagamento_por_Longevidade')

grafico_pagamento_longevidade = px.bar(
    top_departamentos_longevidade,
    x='Nome_Do_Departamento',
    y='Pagamento_por_Longevidade',
    title='Pagamento por Longevidade dos 5 Melhores Departamentos',
)


grafico_pagamento_longevidade.show()
```
